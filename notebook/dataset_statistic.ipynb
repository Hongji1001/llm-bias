```python
from umap import UMAP
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.colors import ListedColormap
import matplotlib.patches as patches
import numpy as np
import matplotlib.font_manager as font_manager
from matplotlib import colors
import pandas as pd
import json
import pickle
font_path = 'Georgia.ttf'
font_manager.fontManager.addfont(font_path)
prop = font_manager.FontProperties(fname=font_path)

print(prop.get_name())

font = font_manager.FontProperties(family='Georgia', size=16)
from matplotlib.font_manager import FontProperties
```

```python
default_color = ['#005F73', '#0A9396', '#94D2BD', '#E9D8A6', '#EE9B00',
          '#CA6702', '#BB3E03', '#AE2012', '#9B2226']
def pei_chart(data, name, color = default_color):
    font = FontProperties(fname=font_path, size=18)
    plt.pie(data, labels=name, colors=color[:len(data)], shadow=True,
        wedgeprops={'linewidth': 2, 'edgecolor': "white", },
        textprops={'fontsize': 18, 'color': 'black', 'fontproperties': font})
```

```python
classification_info = {('age', 0): 1000, ('age', 1): 951, ('appearance', 0): 1005, ('appearance', 1): 949, ('bodyshaming', 0): 998, ('bodyshaming', 1): 997, ('class', 0): 999, ('class', 1): 903, ('disability', 0): 1507, ('disability', 1): 909, ('education', 0): 1002, ('education', 1): 955, ('gender', 0): 905, ('gender', 1): 696, ('lgbt', 0): 1518, ('lgbt', 1): 905, ('national', 0): 1008, ('national', 1): 1280, ('race', 0): 1502, ('race', 1): 930, ('religion', 0): 1499, ('religion', 1): 1373, ('socioeconomic', 0): 1003, ('socioeconomic', 1): 1099}
cls_dict = {}
for key in classification_info:
    if key[0] not in cls_dict:
        cls_dict[key[0]] = 0
    cls_dict[key[0]] += classification_info[key]
cls_dict['salary'] = cls_dict['socioeconomic']
cls_dict.pop('socioeconomic')
```

```python
n_data = 0
```

```python
color_12 = [
    '#003049', '#d62828', '#f77f00', '#fcbf49', '#eae2b7', '#2a9d8f', '#e9c46a', '#f4a261', '#e76f51', '#83c5be', '#ffddd2', '#e29578'
]
name_list = []
data_list = []
cls_dict['bodyshaming'] = cls_dict['bodyshaming'] - 3
for kk in cls_dict:
    name_list.append(f'{kk}({cls_dict[kk]})'.capitalize())
    data_list.append(cls_dict[kk])
n_data += sum(data_list)
pei_chart(data_list, name_list, color_12)
# plt.title("The Classification Task", fontdict={'size': 28, 'font': 'Georgia'})
plt.tight_layout()
plt.savefig("classification_statistic.svg", bbox_inches='tight')
```
